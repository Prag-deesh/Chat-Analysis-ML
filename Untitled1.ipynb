```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Video Games Sales.csv')
```

```python
df
```

```python
df.head()
df.info()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Game Title', y='Global', bins=20, color='red',kde=True)
plt.title('Global Hits')
plt.xlabel('Games')
plt.ylabel('Global Ratings')
plt.show()
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation')
plt.show()
```

```python
total_na = df['North America'].sum()
total_eu = df['Europe'].sum()
total_jpn = df['Japan'].sum()
total_other = df['Rest of World'].sum()
```

```python
regions = ['North America','Europe','Japan','Rest of World']
sales = [total_na,total_eu,total_jpn,total_other]
```

```python
plt.figure(figsize=(10,6))
plt.bar(regions,sales,color=['b','g','b','c'])
plt.title("Total sales by regions")
plt.xlabel('Region')
plt.ylabel('Totale Sales ')
plt.show()
```

```python
yearly_sales = df.groupby('Year')
df['Global'].sum()

plt.figure(figsize=(10,6))
plt.plot(yearly_sales.index,yearly_sales.values,marker='o',linestyle='-')
plt.title('Global trend over the Years')
plt.xlabel('Year')
plt.ylabel('Global in years')
plt.grid=True
plt.xticks(rotation=45)
plt.show()
```
